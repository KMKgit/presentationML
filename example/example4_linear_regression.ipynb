```python
import tensorflow as tf

x_data = [2, 3, 4, 6, 9, 10]        #time
y_data = [40, 40, 45, 80, 100, 100] #score

# shape : 1 array, value: random(between -1.0 to 1.0)
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))

# append placeholder
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# H(x) = Wx + b
hypothesis = W * X + b

# mean
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# a is learning_rate 
a = tf.Variable(0.001)

# minimize cost
optimizer = tf.train.GradientDescentOptimizer(a)
train = optimizer.minimize(cost)

# for init tensor
init = tf.global_variables_initializer()

# session start
with tf.Session() as sess:
    sess.run(init)

    for step in range(10001):       
        sess.run(train, feed_dict={X:x_data, Y:y_data})
        if step % 200 == 0:
            cost_value = sess.run(cost, feed_dict={X:x_data, Y:y_data})
            print (step, cost_value, sess.run(W), sess.run(b))

    print (sess.run(hypothesis, feed_dict={X:[[1], [2], [3]]}))
```
